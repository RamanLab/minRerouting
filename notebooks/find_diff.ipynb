```python
import numpy as np
import pandas as pd
```

```python
model_list = ['e_coli_core', 'iECW_1372', 'iIT341', 'iJO1366', 'iML1515', 'iNJ661', 'iPC815', 'iSSON_1240', 'iYL1228', 'iYS1720', 'STM_v1_0']

for model in model_list:
    zero_norm = pd.read_csv("../results/" + model + "/psl_analysis_zero_norm.csv", index_col=0)
    one_norm = pd.read_csv("../results/" + model + "/psl_analysis_one_norm.csv", index_col=0)
    two_norm = pd.read_csv("../results/" + model + "/psl_analysis_two_norm.csv", index_col=0)

    zero_norm.set_index(["del_rxn1", "del_rxn2"], inplace=True)
    one_norm.set_index(["del_rxn1", "del_rxn2"], inplace=True)
    two_norm.set_index(["del_rxn1", "del_rxn2"], inplace=True)
    
    flag = (zero_norm["sl_size"]!=one_norm["sl_size"])|(zero_norm["sl_size"]!=two_norm["sl_size"])|(two_norm["sl_size"]!=one_norm["sl_size"])
    a = zero_norm.loc[flag]["sl_size"]
    a = pd.DataFrame(a)
    a["one_norm_sl_size"] = one_norm.loc[flag]["sl_size"]
    a["two_norm_sl_size"] = two_norm.loc[flag]["sl_size"]
    a.to_csv("../results/" + model + "/diff.csv")
    display(a)
```
