```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
models = ["iIT341", "iML1515", "iNJ661", "iPC815", "iYL1228", "STM_v1_0"]
columns = ["rxns", "diff", "PathShort", "PathLong", "pathCommon", "solStatus"]
```

```python
def get_dataframes0(name, norm='L0'):
    df = pd.DataFrame()
    for i in columns:
        fname = '../examples/' + name + '/' + name + '_' + i + '_L0.csv'
        fin = open(fname)
        data = fin.read().splitlines()
        fin.close()
        
        df[i] = data[1:]
        df['num_'+i] = df[i].str.split(',').apply(len)
        
    df.loc[df['PathShort']=="", 'num_PathShort'] = 0
    df.loc[df['PathLong']=="", 'num_PathLong'] = 0
    df.loc[df['pathCommon']=="", 'num_pathCommon'] = 0
    df = df.drop(['num_solStatus', 'num_diff'], axis=1)
    df = df.rename(columns={'num_PathShort':'L0_num_PathShort', 
                            'num_PathLong':'L0_num_PathLong', 
                            'num_pathCommon':'L0_num_pathCommon',
                            'num_rxns':'L0_num_rxns'})
    return df

def get_dataframes1(name, norm='L0'):
    df = pd.DataFrame()
    for i in columns:
        fname = '../examples/' + name + '/' + name + '_' + i + '_L1.csv'
        fin = open(fname)
        data = fin.read().splitlines()
        fin.close()
        
        df[i] = data[1:]
        df['num_'+i] = df[i].str.split(',').apply(len)
        
    df.loc[df['PathShort']=="", 'num_PathShort'] = 0
    df.loc[df['PathLong']=="", 'num_PathLong'] = 0
    df.loc[df['pathCommon']=="", 'num_pathCommon'] = 0
    df = df.drop(['num_solStatus', 'num_diff'], axis=1)
    df = df.rename(columns={'num_PathShort':'L1_num_PathShort', 
                            'num_PathLong':'L1_num_PathLong', 
                            'num_pathCommon':'L1_num_pathCommon',
                            'num_rxns':'L1_num_rxns'})
    return df
```

```python
for i in models:
    print("Organism:", i)
    df0 = get_dataframes0(i)
    df1 = get_dataframes1(i)
    display(df1[["L1_num_rxns", "L1_num_PathShort", "L1_num_PathLong", "L1_num_pathCommon"]])
    # df = df0[["L0_num_rxns", "L0_num_PathShort", "L0_num_PathLong", "L0_num_pathCommon"]]
    # df = pd.concat([df, df1[["L1_num_rxns", "L1_num_PathShort", "L1_num_PathLong", "L1_num_pathCommon"]]])
    # display(df)
```
